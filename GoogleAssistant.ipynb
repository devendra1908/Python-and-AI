```python
import pathlib
import textwrap
from gtts import gTTS
import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
from IPython.display import Audio, display


def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
def ask_question(name):
  question = "Hey " + name + ", What do you want ? - "
  question = input(question)
  return question
```

```python
def clasify_questions(question):
  goahead_with_web_search = False
  device_list = ['alarm', 'reminder', 'message', 'call']
  personal_list = ['who are you ?', 'who created you ?']

  device = False
  for i in device_list:
    if i in question.lower():
      device = True

  if device:
    print("This question is releated to device things, which is not support in our google assistant!")

  personal_question = False
  for i in personal_list:
    if i in question.lower():
      personal_question = True

  if device:
    goahead_with_web_search = False
  elif personal_question:
    goahead_with_web_search = False
  else:
    goahead_with_web_search = True

  return goahead_with_web_search
```

```python
def ask_gemini(question):
  modified_prompt = 'Hey give me answer to this question ' + question + ' in maximum of 100 words.'
  response = model.generate_content(modified_prompt)
  to_markdown(response.text)
  return response.text
```

```python
def speak(answer):
  tts = gTTS(answer)
  tts.save('audio.mp3')
  display(Audio('audio.mp3', autoplay=True))
```

```python
from pickle import TRUE
have_any_other_question = 'y'
name = ''

while have_any_other_question == 'y':
  if name == '':
    name = input("Hey, What is your name ? - ")

  question = ask_question(name)

  goahead_with_web_search = clasify_questions(question)
  answer = ''

  if goahead_with_web_search == True :
    answer = ask_gemini(question)

  print(answer)
  speak(answer)

  have_any_other_question = input("\nDo you have any other question ? - ")
```
